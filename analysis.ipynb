```python
import pandas as pd#: Load and Merge Data

trader = pd.read_csv("fear_greed_index .csv")
sentiment = pd.read_csv("fear_greed_index .csv")

trader['date'] = pd.to_datetime(trader['date']).dt.date
sentiment['date'] = pd.to_datetime(sentiment['date']).dt.date

merged = pd.merge(trader, sentiment, left_on="date", right_on="date", how="inner")
```

```python
print(merged.columns)
merged.groupby("classification_x")["value_x"].mean()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x="classification_x", y="value_x", data=merged)
plt.show()
```

```python
sns.barplot(x="classification_x", y="value_x", data=merged)
plt.show()
```

```python
import os

output_dir = "outputs"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

merged.to_csv(os.path.join(output_dir, "results.csv"), index=False)
```
